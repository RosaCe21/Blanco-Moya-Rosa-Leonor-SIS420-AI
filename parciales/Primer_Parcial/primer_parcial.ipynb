cargamos el dataset que vamos a utilizar, en este caso es 'diamonds.csv' que es un datasets con datos sobre diamantes, su peso, corte, color, precio entre otros.

```python
#Importamos el dataset
data = pd.read_csv('./diamonds.csv', delimiter=',')
```

ahora agregamos la columna "RANGO DE PRECIO" de texto que estar en base a la columna de precio (price) donde realizaremos una divisi[on por rangos de precios para agregar un texto consistente y no al azar]

```python
# Agregar una columna de texto aleatorio

conditions = [
    (data['price'] <= 500),         # super economico
    (data['price'] > 500) & (data['price'] < 1500),  # economico
    (data['price'] >= 1500) & (data['price'] < 2500),  # precio normal
    (data['price'] >= 2500) & (data['price'] < 3500),  # costoso
    (data['price'] >= 3500)         # muy costoso
]

# Definir los valores correspondientes a cada condición
choices = ['super economico', 'economico', 'precio normal', 'costoso', 'muy costoso']

# Crear la nueva columna 'economico' basada en las condiciones
data['rango de precio'] = np.select(conditions, choices)
data # Mostrar los primeros datos con la nueva columna
```

agregamos la siguiente columna que se llama CALIDAD DEL CORTE que estara en basre a la columna corte (cut) y asignara un caracter de aucerdo al corte que tenga el diamante

```python
# Agregar una columna de caracter

conditions = [
    (data['cut'] == 'Ideal'),         
    (data['cut'] == 'Premium'),  
    (data['cut'] == 'Good'),  
    (data['cut'] == 'Very Good'),  
    (data['cut'] == 'Fair')         
]

# Definir los valores correspondientes a cada condición
choices = ['i', 'p', 'g', 'v', 'f']

# Crear la nueva columna 'calidad_del_corte' basada en las condiciones
data['calidad_del_corte'] = np.select(conditions, choices)
data
```

Agregamos la ultima columna "PESADO" que esTA en base a la coluna de de peso y en base al peso del diamante asigna un valor boolena ('si','no') dependiendo del peso que tenga

```python
#anadir columna de valores booleanos

data['pesado'] = np.where(data['carat'] > 0.5, True, False)
data
```
